# Exercise Solution
Datasets that will be used are:
* [Iris](https://www.kaggle.com/uciml/iris)
* [Wine](https://www.kaggle.com/uciml/red-wine-quality-cortez-et-al-2009)
* [Company profits](https://www.kaggle.com/datasets/tinasoni/normal-distribution-data)

```python
import pandas as pd

# Load data
wine_data = pd.read_csv("./data/wine/winequality-red.csv", sep=",") # normally-distributed 
iris_data = pd.read_csv("./data/iris/Iris.csv", sep=",") # non-normally distributed
profit_data = pd.read_csv("./data/company_profits/CompanyABCProfit.csv", sep=",") # normally-distributed

def vis_data_box(data, title):
    print('Data set length =', len(data))
    print(data.head())
    # Visualize data
    plt.boxplot(data)
    plt.title(title)
    plt.show()
```

```python
# Wine data
vis_data_box(wine_data['pH'], 'Wine data')
```

```python
# Iris data
vis_data_box(iris_data['SepalLengthCm'], 'Iris sepal length data (in cm)')
```

```python
# Iris data
vis_data_box(profit_data["Profit(Rs '000)"], 'Company profit data')
```

### Answers
**Interpret the visual results and describe them**
* Wine data: There are a few outliers beyond both whiskers.
* Iris data: There are no outliers at all.
* Company profits: There are a few outliers beneath the lower whisker. The data seems to be well distributed around the median of ~ 1000. This can be seen as an indication to being normally distributed.


**How can you test your data for being normally distributed?**
* Histogram: A simple way to visualize the distribution of your data. If the data is normally distributed, the shape of the histogram should resemble the classic bell curve.
* For a normally distributed dataset, about 68% of the data will fall within one standard deviation of the mean, 95% within two standard deviations, and 99.7% within three standard deviations.
* Box Plot: While not a direct test for normality, box plots can provide insights into the symmetry and outliers in your data, which can hint at deviations from normality
